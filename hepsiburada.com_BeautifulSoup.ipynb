```python
from urllib.request import urlopen
import requests
from bs4 import BeautifulSoup as bts
import pandas as pd
import re
import time
import sys
from datetime import datetime
from datetime import date
```

```python
def getAndParseURL(url):
    result = requests.get(url,headers={"User-Agent":"Mozilla/5.0"})
    soup = bts(result.text, 'html.parser')
    return soup
```

```python
getAndParseURL('https://www.hepsiburada.com/laptop-notebook-dizustu-bilgisayarlar-c-98')
```

```python
PAGE_NUM = []
def find_page_num():
    

    html = getAndParseURL("https://www.hepsiburada.com/laptop-notebook-dizustu-bilgisayarlar-c-98")
    
    for i in range(0,50):
 
        PAGE_NUM.append("https://www.hepsiburada.com/laptop-notebook-dizustu-bilgisayarlar-c-98"+str("?sayfa=")+str(i))
        print(PAGE_NUM)

    return PAGE_NUM
find_page_num()
```

```python
ALL_PRODUCT = []
def all_product():
    
    for i in PAGE_NUM[::]:
        html = getAndParseURL(i)
        for s in html.findAll("div",{"data-test-id":"product-card-container"}):
            ALL_PRODUCT.append(s.a["href"])

    return ALL_PRODUCT

all_product()
```

```python
result = []
for pr in ALL_PRODUCT[:]:
    
    try:        
        prod = getAndParseURL(pr)
        #PRODUCT_UNQ=pr.split("-")[-1]
    except:
        continue
        
    PRODUCT_NAME=prod.find("span",{"class":"product-name"}).text
    
    PRICE=prod.find("span",{"id":"offering-price"})["content"]
    
    BRAND_NAME=prod.find("span",{"class":"brand-name"}).text.strip()
    try:
        RATING_STAR=prod.find("span",{"class":"rating-star"}).text.strip()
    except:
        RATING_STAR=0
    try:
        NUMBERofCOMMENTS=prod.find("div",{"id":"comments-container"}).span.text
    except:
        NUMBERofCOMMENTS=0
    DATE=date.today()
  
    result.append([PRODUCT_NAME, PRICE, BRAND_NAME,RATING_STAR,NUMBERofCOMMENTS,DATE])
    print(pr)

columns = ["product_name","price","brand_name","rating_star","Number_of_Comments","Date"]
dfMatch = pd.DataFrame.from_records(result, columns=columns)
```
