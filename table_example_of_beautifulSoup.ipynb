```python
from bs4 import BeautifulSoup as bts
import requests
```

```python
url="https://www.tcmb.gov.tr/wps/wcm/connect/TR/TCMB+TR/Main+Menu/Istatistikler/Enflasyon+Verileri/"

html=requests.get(url)
soup=bts(html.text,'html.parser')
soup.findAll('tr',limit=2)

headers=[th.getText() for th in soup.findAll('tr',limit=2)[0].findAll('th')]
headers
```

```python
rows=soup.findAll('tr')[1:]
stats=[[td.getText() for td in rows[i].findAll('td')]
       for i in range(len(rows))]
```

```python
stats
```

```python
import pandas as pd

stats_df=pd.DataFrame(stats,columns=headers)
```
