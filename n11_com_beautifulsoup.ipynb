```python
def getAndParseURL(url):
  result=requests.get(url,headers={"User-Agent":"Mozilla/5.0"})
  soup=bts(result.text, 'html.parser')
  return soup
```

```python
html=getAndParseURL("https://www.n11.com/bilgisayar?m=Lenovo")

for i in html.findAll("div",{"class":"pagination"}):
  for b in i.findAll("a"):
    print(b["href"])
```

```python
PAGE_NUM = []
def findPage(url):
  html=getAndParseURL(url)

  for i in html.findAll("div",{"class":"pagination"}):
    for b in i.findAll("a"):
      PAGE_NUM.append(b["href"])
```

```python
findPage("https://www.n11.com/bilgisayar?m=Lenovo")
```

```python
URUN_LINK=[]
def productLink(url):
  for sayfa in NEW_PAGE_NUM:
    html=getAndParseURL(sayfa)
    
    for urun in html.findAll("div",{"class":"pro"}):
      URUN_LINK.append(urun.a.get("href"))
```

```python
import re
RESULT = []
for sonuc in URUN_LINK:
  try:
    html=getAndParseURL(sonuc)
    ISIM= html.find('h1',{"class":"proName"}).text.strip()
    SATICI_ISMI=html.find("a",{"class":"main-seller-name"})

    script= html.findAll("script")[2]
    string= str(script)
    start= string.find("lowPrice")+12

    end= int(start)+13
    answer=string[start:end]

    result=(re.search(r'\d+',answer))
    FIYAT=result[0]
    STOK=html.find("input",{"class","skuStock"}).get("value")

    star=html.find("div",{"class","ratingCont"}).span
    starResult=str(star)

    result= int (re.search(r'\d+',starResult)[0])
    YILDIZ=result/20

    RESULT.append([ISIM,SATICI_ISMI,FIYAT,STOK,YILDIZ])


  except:
    continue

columns= ["isim","satici","fiyat","stok","yildiz"]

df= pd.DataFrame.from_records(RESULT,columns=columns)

df
```
